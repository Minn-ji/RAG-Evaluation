```python
from BLEU import bleu
import pandas as pd
from datasets import load_dataset
from pathlib import Path
from datasets import Dataset

PATH = Path('.').resolve().parent.parent / 'data'
EXAMPLE_DATASET=PATH / "results/03-01_bmt_result.csv"
ANS_DATASET= PATH / "new_rag-eval-ko-dataset-public.csv"
print(EXAMPLE_DATASET)
df = pd.read_csv(EXAMPLE_DATASET)
df2 = pd.read_csv(ANS_DATASET)
```

```python
df
```

```python
questions = df["question"]
questions[0]
```

```python
reference = df.loc[:,["retrieved_cont1", "retrieved_cont2", "retrieved_cont3", "retrieved_cont4", "retrieved_cont5"]]
reference = reference.values.tolist()
reference[0]
```

```python
target_answer = df2["target_answer"]
target_answer = target_answer.values.tolist()
target_answer[0]
```

```python
from ragas.metrics import BleuScore, faithfulness
from ragas import SingleTurnSample
from ragas.metrics import RougeScore

import numpy as np

from ragas import evaluate

# Example data
dataset_list = []
for q, r, a in zip(questions, reference, target_answer):
    dataset_list.append(
        # SingleTurnSample(user_input=q, response=a, reference=",".join(r))
        SingleTurnSample(user_input=q, response=a, retrieved_contexts=r)
    )

scorer = BleuScore()
# Convert the data to a Hugging Face Dataset
# dataset = Dataset.from_list(dataset_list)

print(dataset_list)
```

```python
dataset_list[0]
```

```python
results = []
for i in dataset_list:
    # 'await' tells Python to run the coroutine and wait for its result
    result = await scorer.single_turn_ascore(i) 
    results.append(result)
```
